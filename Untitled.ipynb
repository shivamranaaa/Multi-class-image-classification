## Import Requried Libraries.  

```python
# Importing the best Pretrained models for these kind of applications.
from tensorflow.keras.applications import MobileNetV2, VGG16, VGG19

# Model
from sklearn.ensemble import RandomForestClassifier

# For checking out the proformace of model.
from sklearn.metrics import classification_report,confusion_matrix

# To do operations on images.
import cv2

# For Extracting the Images from Local drive. 
import os

# For mathematical  opertions.
import numpy as np

# For visualization.
import matplotlib.pyplot as plt
import seaborn as sns

# To ingore warnings.
import warnings
warnings.filterwarnings("ignore")
```

```python
# Some important and required parameters.
Output_labels=103
size=100
```

```python
# Extracting the train data and it's requried labels.
train=[]
train_label=[]
for i in os.listdir("Fruit-Images-Dataset-master/Training"):
    for j in os.listdir(os.path.join("Fruit-Images-Dataset-master/Training",i))[:40]:#Extracting first 40 images bcoz of MemoryError.
        img=cv2.imread(os.path.join("Fruit-Images-Dataset-master/Training",i,j))
        img=cv2.resize(img,(size,size))
        train.append(img)
        train_label.append(i)
```

```python
# Extracting the test data and it's requried labels.
test=[]
test_label=[]
for i in os.listdir("Fruit-Images-Dataset-master/Test"):
    for j in os.listdir(os.path.join("Fruit-Images-Dataset-master/Test",i))[:5]:#Extracting first 5 images bcoz of MemoryError.
        img=cv2.imread(os.path.join("Fruit-Images-Dataset-master/Test",i,j))
        img=cv2.resize(img,(size,size))
        test.append(img)
        test_label.append(i)
```

```python
# Converting the list to arrays and scaling or normalizing the data by dividing with 255.0
train=np.array(train)/255.0
train_label=np.array(train_label)

test=np.array(test)/255.0
test_label=np.array(test_label)
```

```python
# Converting the Categorical labels into numeric form.
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
train_label=le.fit_transform(train_label)
test_label=le.fit_transform(test_label)
```

```python
# Trying to make it easily understandable to Anyone.
X_train,y_train,X_test,y_test=train , train_label , test , test_label
```

```python
# Converting into one hot form to the train and test data.
from tensorflow.keras.utils import to_categorical
one_hot_train=to_categorical(y_train)
one_hot_test=to_categorical(y_test)
```

```python
# Pretrained model VGG16.
vgg=VGG16(include_top=False,input_shape=(size,size,3))
```

```python
for i in vgg.layers:
    i.trainable=False # I dont want to try the whole model thats why i used it.
vgg.summary()
```

```python
feature_Extraction=vgg.predict(X_train) # Extracting the features from the train Images.
```

```python
rf_X_train=feature_Extraction.reshape(feature_Extraction.shape[0],-1) # Converting that features into the input format of ml models.
```

```python
# Model.
rf=RandomForestClassifier(n_estimators = 50, random_state = 42) # n_estimators=50 found by help of hyperparameter tuning.
rf.fit(rf_X_train,y_train)
```

```python
feature_Extraction_X_test=vgg.predict(X_test) # Extracting the features from the test Images.
rf_X_test=feature_Extraction_X_test.reshape(feature_Extraction_X_test.shape[0],-1) # Converting that features into the input format of ml models.
```

```python
y_pred=rf.predict(rf_X_test) # Predicting the result of the test data.
y_pred
```

```python
plt.rcParams["figure.figsize"]=40,40
sns.heatmap(confusion_matrix(y_pred,y_test),annot=True) # Plot the Confusion matrix for better visualization.
```

```python
# To know about the Recall, Presion and F1 score.
print(classification_report(y_pred,y_test)) # Avg Recall, Presion and F1 score is 97.
```
